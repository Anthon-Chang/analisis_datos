```python
import pandas as pd
import numpy as np
```

```python
# Reemplaza 'ruta/dataset.csv' con la ruta a tu archivo
df = pd.read_csv('LigaEspanola2023-2024-Resultados.csv')

# Vista previa de los primeros registros
df
```

```python
# Tamaño del dataset (filas y columnas)
print(df.shape)

# Ver información general sobre las columnas, tipos de datos y valores faltantes
print(df.info())

# Estadísticas básicas (solo numéricas por defecto)
print(df.describe())

```

```python


# Eliminar filas con valores faltantes
df = df.dropna()

# Eliminar duplicados
df = df.drop_duplicates()

# Crear una columna incremental y convertirla en índice
df['Indice'] = range(1, len(df) + 1)  # Índices desde 1 hasta el número de filas
df.set_index('Indice', inplace=True)


# Guardar el dataset limpio
#df.to_csv('C:/Users/User/Desktop/analisis/ligaEspañola_limpio.csv')
df.to_csv("C:/Users/User/Desktop/analisis/ligaEspañola_limpio.csv", encoding="utf-8")
```
